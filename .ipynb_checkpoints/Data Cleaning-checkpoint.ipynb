## Context
#### Mumbai is the capital city of the Indian state of Maharashtra. It is a major financial and cultural hub in India, with a population of over 20 million people. The real estate market in Mumbai is known to be one of the most expensive in India. This house price dataset for Mumbai contains information on the sale prices of residential properties in the city, such as apartments and houses. It also includes information on the location, size, and age of the properties.

## Columns
### bhk
#### contains the number of bedroom, hall, kitchen collectively.

### type
#### contains the type of house, type can be one of the following:

####  • apartment
####  • villa
#### • independent house
#### • studio apartment

#### Note: the majority of data is based on one of the above types.

### locality
#### contains information about the locality of house.

### area
#### contains the area of house, unit of measurement is sq ft.

### price
#### contains the price value for the house.

### price_unit
#### contains the price unit for the house which can be from given below:

#### • L -> represents Lakh (1 Lakh = 1,00,000 and 10 Lakhs = 1 Million)
#### • Cr -> represents Crore (1 Crore = 1,00,00,000 and 10 Million = 1 Crore)

#### Note: the currency of price here is Indian Rupee.

### region
#### contains the region of the house.

### status
#### contains information about the status of the house which can be either of given below:

#### • Ready to move -> the house is ready to use
#### • Under Construction -> the house is currently under construction

### age
#### contains the information regarding age of house which can any of below:

#### • New: a new house for sale
#### • Resale: an old house for resale

#### Note: the ‘unknown’ values in age represents that no data available for that instance, unknown is placeholder for null values in age.

## 1. Import Necessary Libraries

```python
import re
import string
import pandas as pd
```

## 2. Import Dataset

```python
data = pd.read_csv("Datasets\Raw_Property.csv")
data
```

```python
data.shape
```

## 3. Data Preprocessing

### 3.1 Remove the Unwanted Symbols and text

```python
data["Property_Age"] = data["Property_Age"].str.replace(' Old', '')
data
```

```python
data["Rate_SqFt"] = data["Rate_SqFt"].str.replace(',', '')
```

```python
data["Availability"].unique()
```

```python
data["Availability"] = data["Availability"].str.title()
data["Availability"] = data["Availability"].str.replace(' Property', '')
data["Availability"] = [i.lstrip() for i in data["Availability"]]
```

```python
data["Availability"].unique()
```

### 3.2 Set Column Area in SqFt

```python
area = []
for i in range(2581):
    clean_sqft = re.sub('[^0-9.]', " ", data["Area_Tpye"][i])
    area.append(clean_sqft.split()[0])
```

```python
data["Area_SqFt"] = area
```

### 3.3 Set Column Type of Carpet Area

```python
carpet = []
for i in range(2581):
    clean_carpet = re.sub('[^a-zA-Z]', " ", data["Area_Tpye"][i])
    carpet.append(clean_carpet.split()[0] + ' ' + clean_carpet.split()[1])
data["Area_Tpye"] = carpet
data
```

```python
data["Area_Tpye"] = data["Area_Tpye"].str.title()
```

```python
data.Area_Tpye.unique()
```

```python
data["Area_Tpye"] = data.Area_Tpye.str.replace('Super Built', 'Super Built Up')
data["Area_Tpye"] = data.Area_Tpye.str.replace('Built Up', 'Built Up Area')
data["Area_Tpye"] = data.Area_Tpye.str.replace('Carpet Area ','Carpet Area')
```

```python
data.Area_Tpye.unique()
```

### 3.4 Remove Unwanted Text from Floor No

```python
data["Floor_No"] = data["Floor_No"].str.replace('Ground', '0')
data["Floor_No"] = data["Floor_No"].str.replace('Basement', '-1')
```

```python
floor= []
for i in range(len(data)):
    clean_floor = re.sub('[^0-9.]', " ", data["Floor_No"][i])
    floor.append(clean_floor)
data["Floor_No"] = floor
data
```

```python
data['Location'].unique()
```

```python
data['Location'] = [i.lstrip() for i in data['Location']]
data['Location'].unique()
```

### 3.5 Set Region Column

```python
location = []
for i in range(2581):
    clean_location = re.sub('[^a-zA-Z]', " ", data['Location'][i])
    location.append(clean_location)
data['Region'] = location
```

```python
data['Region'] = data['Region'].str.title()
words = ['[0-9]', 'East','West', 'South', 'Suburbs', 'Sector', 'Beyond', 'And Beyond', 'Scheme']
data['Region'] = data['Region'].str.replace('|'.join(words), "", regex = True).str.strip()
```

```python
location = []
for i in range(2581):
    try:
        location.append(data['Region'][i].split()[-3] + ' ' + data['Region'][i].split()[-2])
    except:
        location.append(data['Region'][i].split()[-2] + ' ' + data['Region'][i].split()[-1])
        
data['Region'] = location
data
```

```python
data.Region.value_counts().head(30)
```

```python
add = []
for i in range(2581):
    clean_add = re.sub('[^a-zA-Z0-9]', " ", data['Location'][i])
    add.append(clean_add)
data['Location'] = add

data['Location'] = data['Location'].str.replace('   ', ' ')
data['Location'] = data['Location'].str.replace('  ',' ')
```

```python
data
```

### 3.6 Replace the all Values in Lac's from Price Column

```python
def convertor(x):
    if 'Lac' in x:
        return f"{(float(x.strip('Lac')) * 1):,.1f}"
    elif 'Crore' in x:
        return f"{(float(x.strip('Crore')) * 100):,.1f}"
    
data['Price_Lakh'] = data['Price'].apply(convertor)
data['Price_Lakh'] = data['Price_Lakh'].str.replace(',', '')
```

```python
data.isna().sum()
```

```python
data.dropna(inplace=True)
data.reset_index(drop=True, inplace=True)
```

### 3.8 Sort the all Columns

```python
data = data[['Property_Name','Location','Region','Property_Age','Availability','Area_Tpye','Area_SqFt','Rate_SqFt','Floor_No','Bedroom','Bathroom','Price_Lakh']]
data
```
